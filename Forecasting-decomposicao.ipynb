```python
import numpy as np
import matplotlib.pyplot as plt

from sktime.datasets import load_airline
```

```python
y = load_airline()
```

```python
y.plot(title='Vendas de passagens')
```

```python
y.rolling(12).mean().plot(title='Tendncia estimada')
```

```python
(y - y.rolling(12).mean()).plot(title='Vendas de passagens')
```

```python
(y / y.rolling(12).mean()).plot(title='Vendas de passagens')
```

```python
y[list(range(0,y.shape[0],12))]
```

```python
import numpy as np

season = np.zeros(12)
for i in range(12):
    season[i] = y[list(range(i,y.shape[0],12))].mean()
```

```python
plt.plot(season)
```
