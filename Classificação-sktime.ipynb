```python
import numpy as np
import pandas as pd

from sktime.transformations.panel.rocket import Rocket
from sklearn.linear_model import RidgeClassifierCV
```

Leitura do conjunto de dados

```python
cbf_train = pd.read_csv('CBF_TRAIN.tsv', delimiter='\t', header=None)
cbf_train_labels = cbf_train[0].to_numpy()
cbf_train = cbf_train.drop(0, axis = 1).to_numpy()
```

```python
cbf_test = pd.read_csv('CBF_TEST.tsv', delimiter='\t', header=None)
cbf_test_labels = cbf_test[0].to_numpy()
cbf_test = cbf_test.drop(0, axis = 1).to_numpy()
```

Fazer a transformação ROCKET e classificar.

```python
rocket = Rocket()  # by default, ROCKET uses 10,000 kernels
rocket.fit(cbf_train)
cbf_train_transform = rocket.transform(cbf_train)
```

```python
classifier = RidgeClassifierCV()
classifier.fit(cbf_train_transform, cbf_train_labels)
```

```python
cbf_test_transform = rocket.transform(cbf_test)
classifier.score(cbf_test_transform, cbf_test_labels)
```
